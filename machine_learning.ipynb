OpenCV 中的机器学习
===========

基本概念
=======
1. 什么是机器学习
    OpenCV中的机器学习只涉及统计性机器学习,机器学习还包含其他方面,如贝叶斯网络,马尔可夫随机场,图模型等.
    推荐学习资料:
    
2. 分类器:
    机器学习的过程是,先通过原始数据训练分类器(相当于人脑获取知识的过程), 建立某种预测模型.   这些用于训练的数据就是训练集. 分类器具备一定的"知识"后,需要用一定的数据来测试这个分类器(大脑). 测试通过后,需要一些现实数据来验证分类器的效果.这需用于验证的数据就叫做验证集.
    测试集不能用来训练分类器.
    
3. 有监督数据和无监督数据
    如果数据有标签,如年龄,对这类数据的学习就是有有监督的学习. 如果数据没有标签,对这类数据的学习就是无监督的学习.

4. 分类算法和聚类算法
    对于有标签的数据,如姓名,年龄. 对这些数据的学习目的是将这些数据按姓名,年龄进行分类. 这类学习过程叫分类算法.
    对于无标签的数据, 即上面的无监督学习算法,目的是将这些无标签数据按照他们的距离(相似性)远近归类, 这类学习过程叫聚类算法(clustering algorithm)
    
5. 产生式模型和鉴别式模型
    机器学习算法的两个基本原理是统计学和概率,OpenCV主要支持基于统计学的算法, 基于统计学开发的算法主要倾向于鉴别式算法,即通过给定数据来判断类别.  基于概率的算法主要用于根据给定的特征生成数据分布.鉴别式算法在根据给定数据做预测上有优势,而产生式算法的优势在于提供强大的数据表达或者有条件地生成新数据.

    公式:  
        在二维和三维空间中的欧氏距离的就是两点之间的距离
        二维的公式
        d = sqrt((x1-x2)^2+(y1-y2)^2)
        三维的公式
        d=sqrt((x1-x2)^2+(y1-y2)^2+(z1-z2)^2)

概率论基础知识
==
1.  概率
        经过多次实验得到随机事件出现的频率,当实验次数充分多时,事件A的频率无限接近一个数值p,这个数值p就是事件的概率.
        在相同条件下进行n次重复试验，如果随机事件A发生的次数为m，那么m/n称为随机事件A的频率（frequency）；当试验重复数n逐渐增大时，随机事件A的频率越来越稳定地接近某一数值p，那么就把p称为随机事件A的概率。这样定义的概率称为统计概率（statistics probability），或者称后验概率（posterior probability）。
        机事件的概率p是不可能准确得到的。通常以试验次数n充分大时随机事件A的频率作为该随机事件概率的近似值:
        P（A）=p≈m/n（n充分大）
        
2.  古典概型(又称先验概率)
        古典概型是在通用概率的基础上增加了一些条件得到的概率模型, 这些条件是:
        a. 实验的所有可能的结果是有限个,即样本空间中的基本事件个数为有限个.
        b. 各个实验的可能结果出现的可能必相等.即所有基本事件的发生是等可能的.
        c. 实验的所有可能结果两两不相容.
    
        古典概型的概率计算:
        P(A) = m/n
    
3.  随机变量: 
        作一次实验,其结果有多种可能,每一种可能结果都可用一个数来表示,把这些数作为变量X的取值范围,则实验结果可用变量X 来表示.
        如果表示试验结果的变量x, 其可能值至多为可数个. 且以各种确定的概率取这些不同的值,则称x为离散随机变量.如果其可能取值为某范围内的任何数值,且x在取值范围内的任何一个区间中取值时,其概率是确定的,则称x为连续随机变量.
            
3.1 离散随机变量的概率分布
        如果将离散随机变量x的一切可能取值Xi(i=1,2,...)及其对应的概率pi,记着:
        P(x=xi)=pi i=1,2,...
        则称该式为离散随机变量的x的概率分布. 常用分布列来表示离散随机变量:
        x1 x2 ... xn ...
        
        p1 p2 ... pn ...
