OpenCV 中的机器学习

基本概念
=======
1. 什么是机器学习
    OpenCV中的机器学习只涉及统计性机器学习,机器学习还包含其他方面,如贝叶斯网络,马尔可夫随机场,图模型等.

2. 分类器:
    机器学习的过程是,先通过原始数据训练分类器(相当于人脑获取知识的过程), 建立某种预测模型.   这些用于训练的数据就是训练集. 分类器具备一定的"知识"后,需要用一定的数据来测试这个分类器(大脑). 测试通过后,需要一些现实数据来验证分类器的效果.这需用于验证的数据就叫做验证集.
    测试集不能用来训练分类器.
    
3. 有监督数据和无监督数据
    如果数据有标签,如年龄,对这类数据的学习就是有有监督的学习. 如果数据没有标签,对这类数据的学习就是无监督的学习.
