在数据中搜索模式是一个很基本的问题,已经有很长的成功历史. 例如,在16世纪,丹麦天文学家第谷(Tycho Brahe)作了大量的天文观测, 德国天文学家开普勒在此基础上发现了行星运行的经验性规律,这一规律反过来又为经典力学的发展提供了基础.同样,在二十世纪早期,原子光谱规律的发现对量子物理的发展和验证过程中扮演了重要角色.模式识别就是在数据中应用计算机算法自动发现规律,并运用这些规律作出一些行为,如将数据进行分类等.

### 1.2 概率论

有两个盒子,一个红的,一个蓝的,在红色盒子中,有2个苹果和6个桔子.蓝色盒子中,有3个苹果和一个桔子.随便从一个盒子中,随机拿一个水果,观察我们拿到了哪种水果,然后放回原盒子.这样重复多次.假设我们对两个盒子做了100次实验,红色盒子拿了40次,蓝色盒子拿了60次. 
    
哪个盒子被选中,这是一个随机的结果,可能是红色盒子,也可能是蓝色盒子,这是一个变量. 就像常见的变量x,y,z. x,y,z的取值通常是实数或自然数,或其他规定的值域.同理,用来表示随机结果的变量也是变量, 我们叫它随机变量,这个变量的值域是所有基本事件. 例如,选择哪个盒子的随机变量B的值域是红色盒子或蓝色盒子.记为B(r,b),同理,选中什么水果可用随机变量F表示.值域为(a,o),对应(apple,orange)
    
问题:
- 选中苹果的概率是多少.
- 假设选中了一个桔子,那么选中蓝色盒子的概率是多少.

这些问题都是模式识别中相关的问题. 要回答这类问题,需要掌握概率的基本规则: "和规则"与"积规则"

为了从上述概率派生出更通用的概率理论, 假定有两个随机变量X,值域$\{x_i\}$. 其中 $_i=1,...,M$, 和Y, 值域$\{y_j\}$. 其中 $_j=1,...,L$. 假定M=5,L=3, 假定所有这些变量个数为N, 我们将 $X=x_i, Y=y_j$ 的处的实例个数表示为 $ n_{ij}$. 这里的实例个数指的是试验结果的概率.比如,将上面的 选中盒子随机变量B和选中苹果的随机变量F 分别对应这里的X,Y, 就可以将实例个数理解成是选中苹果并且选中蓝色盒子的概率.

X取值$x_i$ 并且Y取值$y_j$的概率公式记作
$$p(X=x_i, Y=y_j)=\frac{n_{ij}}{N}$$ 

这里有个隐式条件,$\lim {N \to \infty}$, 同样,不考虑Y的值,X取$x_i$的概率记作p(X = $x_i$). 用分数表示就是第i列的所有点:
$$ p(X = x_i) = \frac{c_i}{N} $$

因为在上图中,第i列的实例数就是这一列所有单元的和,所以有 $c_i=\sum_j{n_{ij}}$

于是:
$$
p(X = x_i) = \sum_{j=1}^{L} {p(X=x_i, Y=y_i)}
$$

这就是概率的和规则,这个概率$p(X = x_i)$也叫边际概率,因为它排斥了其他的变量(这里指的是Y)

如果只考虑$X=x_i$的实例,那么以X变量取值作为条件的Y变量 $Y=y_i$ 的概率$p(Y=y_i|X=x_i)$叫做当$X=x_i$时Y的条件概率
