1.2 概率论
    有两个盒子,一个红的,一个蓝的,在红色盒子中,有2个苹果和6个桔子.蓝色盒子中,有3个苹果和一个桔子.随便从一个盒子中,随机拿一个水果,观察我们拿到了哪种水果,然后放回原盒子.这样重复多次.假设我们对两个盒子做了100次实验,红色盒子拿了40次,蓝色盒子拿了60次. 
    
    哪个盒子被选中,这是一个随机的结果,可能是红色盒子,也可能是蓝色盒子,这是一个变量. 就像常见的变量x,y,z. x,y,z的取值通常是实数或自然数,或其他规定的值域.同理,用来表示随机结果的变量也是变量, 我们叫它随机变量,这个变量的值域是所有基本事件. 例如,选择哪个盒子的随机变量B的值域是红色盒子或蓝色盒子.记为B(r,b),同理,选中什么水果可用随机变量F表示.值域为(a,o),对应(apple,orange)
    
    问题: 
    1. 选中苹果的概率是多少.
    2. 假设选中了一个桔子,那么选中蓝色盒子的概率是多少.
    
    这些问题都是模式识别中相关的问题. 要回答这类问题,需要掌握概率的基本规则: "和规则"与"积规则"
